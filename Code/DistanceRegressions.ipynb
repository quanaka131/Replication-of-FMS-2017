```python
# ============================================================
# OLS with COUNTY & YEAR FIXED EFFECTS using 100-mile exposure
# ============================================================
# Model:
#   Δy_it = β1 * d100newfossilval_cap_it
#         + β2 * d100newfossilval_cap_{i,t-1}
#         + α_i (county FE) + ω_t (year FE) + ε_it
#
# Variables (100-mile “radius” panel):
# - d100newfossilval_cap : value of *new* oil+gas within 100 miles of county i in year t,
#                          per person, in **MILLIONS of 2014 USD**.
#   -> A change of 0.001 equals **$1,000 per person**.
# - d_d100irswag_cap     : change in IRS wage income per person aggregated within 100 miles.
# - d_d100wages_cap      : change in BLS wages per person aggregated within 100 miles.
#
# Fixed effects & SEs:
# - α_i and ω_t soak up time-invariant county traits and national/year shocks.
# - Two-way clustered SEs (by county and by year) allow serial & spatial correlation.
#
# Requirements: `pip install linearmodels`
# Input: `df_stat` from your earlier “100-mile” sample construction (balanced panel).
# ============================================================

import numpy as np
import pandas as pd
from linearmodels.panel import PanelOLS

# ---------- 1) Select and sanitize the 100-mile panel ----------
# Keep only IDs, the 100-mile shock, and the two outcomes
need = ["fips", "year", "d100newfossilval_cap", "d_d100irswag_cap", "d_d100wages_cap",  "d_d100irsoth_cap"]
df100 = df_stat[need].copy()  # df_stat came from your radius=100 prep

# Ensure numeric dtypes (guards against object/strings from import)
for c in need[2:]:
    df100[c] = pd.to_numeric(df100[c], errors="coerce")

# Sort by county-year and build the **within-county lag** of the 100-mile shock
df100 = df100.sort_values(["fips", "year"])
df100["L1_d100newfossilval_cap"] = (
    df100.groupby("fips")["d100newfossilval_cap"].shift(1)
)

# Set MultiIndex so PanelOLS recognizes the panel structure
panel = df100.set_index(["fips", "year"])

# Regressor matrix used in both outcome equations
X = panel[["d100newfossilval_cap", "L1_d100newfossilval_cap"]]

# ---------- 2) Helper to run one FE regression and print a compact summary ----------
def run_radius_fe(y_col: str, label: str):
    """
    y_col : outcome column name (e.g., 'd_d100irswag_cap')
    label : pretty label for printing
    """
    # Join outcome with regressors and drop incomplete rows
    d = panel[[y_col]].join(X).dropna()

    # Context counts
    n_entities = d.index.get_level_values(0).nunique()  # unique counties
    n_years    = d.index.get_level_values(1).nunique()  # unique years

    # Two-way FE: county FE (entity_effects) + year FE (time_effects)
    mod = PanelOLS(
        d[y_col],
        d[["d100newfossilval_cap", "L1_d100newfossilval_cap"]],
        entity_effects=True,   # α_i
        time_effects=True      # ω_t
    )

    # Fit with two-way clustered SEs (county & year)
    res = mod.fit(
        cov_type="clustered",
        cluster_entity=True,
        cluster_time=True
    )

    # Header and compact coefficient table
    print(f"\n=== OLS FE (county & year) — {label} ===")
    print(f"Observations (county–years): {res.nobs:,}  |  Counties: {n_entities:,}  |  Years: {n_years}")
    out = pd.DataFrame({
        "Coefficients": res.params,                 # β estimates
        "SE (2-way Clustered)": res.std_errors,
        "t-stat": res.tstats,
        "p-value": res.pvalues
    })
    display(out.loc[["d100newfossilval_cap", "L1_d100newfossilval_cap"]].round(4))

    # R² diagnostics:
    #  - within: variance explained after removing county and year means
    #  - overall: includes fixed effects
    r2w = getattr(res, "rsquared_within", np.nan)
    r2o = getattr(res, "rsquared", np.nan)
    print(f"R2 (within): {r2w:.3f} | R2 (overall): {r2o:.3f}")

    # Interpretation tip:
    # Multiply a coefficient by **0.001** to get the effect **per $1,000/person** of new 100-mile production.
    return res

# ---------- 3) Run both outcomes ----------
res_irs_100 = run_radius_fe("d_d100irswag_cap", "IRS wage income per person (Δ), 100-mile exposure")
res_bls_100 = run_radius_fe("d_d100wages_cap",  "BLS wage bill per person (Δ), 100-mile exposure")
res_irs_other_100 = run_radius_fe(
    "d_d100irsoth_cap",
    "IRS other income (royalties & nonwage business) per person (Δ), 100-mile exposure"
)
```

```python
# =============================================================================
# Distance-IV (100-mile exposure) — 2SLS with COUNTY & YEAR FE (absorbed)
#
# WHAT THIS DOES
#  - Estimates the causal effect of new O&G production within 100 miles on income
#    changes using instrumental variables, while absorbing county and year fixed
#    effects via two-way demeaning and clustering SEs by county and year.
#
# INPUT
#  - df_stat2: one row per county–year, already filtered/balanced (e.g., 2005–2012)
#       IDs:          fips, year
#       Shock:        d{R}newfossilval_cap,  L1_d{R}newfossilval_cap
#       Instruments:  d{R}newfossilval_cap_ins,  L1_d{R}newfossilval_cap_ins
#       Outcomes:     d_d{R}irswag_cap, d_d{R}wages_cap, d_d{R}irsoth_cap
#
# UNITS & INTERPRETATION
#  - Shock is in MILLIONS of 2014 USD per person. To read “per $1,000 per person”,
#    multiply any coefficient by 0.001.
#
# METHOD
#  - Two-way demeaning of y, X (endogenous), and Z (instruments) absorbs FE.
#  - 2SLS (IV) with two-way clustered SEs (county, year).
#
# REQUIREMENTS
#  - pip install linearmodels
# =============================================================================

import numpy as np
import pandas as pd
from linearmodels.iv import IV2SLS

RADIUS = 100  # miles

## Two-way demeaning helper: Removes county mean, year, mean, and adds back grand mean
    # County and year FE are removed so we can run IV w/o explicitly building dummy matrices
def tw_demean(d: pd.DataFrame, cols, ent="fips", tim="year") -> pd.DataFrame:
    """
    Two-way demeaning to absorb county & year FE:
      x_it - mean_i(x) - mean_t(x) + grand_mean(x)
    """
    out = pd.DataFrame(index=d.index)
    for c in cols: 
        v = d[c]
        out[c] = v - d.groupby(ent)[c].transform("mean") \
                     - d.groupby(tim)[c].transform("mean") \
                     + v.mean()
    return out

## IV Runner (requires current and lag instruments

def run_distance_iv_strict(df_in: pd.DataFrame, y_col: str, label: str, R=RADIUS):
    # Column names
    shock   = f"d{R}newfossilval_cap"
    shockL  = f"L1_d{R}newfossilval_cap"
    ins     = f"d{R}newfossilval_cap_ins"
    insL    = f"L1_d{R}newfossilval_cap_ins"

    # Require all columns
    need = ["fips","year", y_col, shock, shockL, ins, insL]
    missing = [c for c in need if c not in df_in.columns]
    if missing:
        raise ValueError(f"df_in missing required columns: {missing}")

    # Keep and drop true missings (zeros are allowed): Only rows with complete y,X,Z are kept. Zeros fine; NaN removed.
    use = df_in[need].dropna().copy()
    if use.empty:
        raise ValueError("No rows after aligning y, X, Z. Check lags/instruments/sample window.")

    # Absorb FE (demean y, X, Z consistently): Same deamning transofmraiton is applied to y, X, AND Z.
        # In IV, regressors and instruments must lived in the same transformed space.

    y_dm = tw_demean(use, [y_col])[y_col]
    X_dm = tw_demean(use, [shock, shockL])
    Z_dm = tw_demean(use, [ins,   insL])

    # Final NA guard post-transform
    keep = (~y_dm.isna()) & (~X_dm.isna().any(axis=1)) & (~Z_dm.isna().any(axis=1))
    y_dm, X_dm, Z_dm = y_dm[keep], X_dm.loc[keep], Z_dm.loc[keep]
    ids = use.loc[keep, ["fips","year"]]

    # Two-way clustered SEs (county, year)
    clusters = np.column_stack([ids["fips"].to_numpy(), ids["year"].to_numpy()])

    # 2SLS (no intercept needed after demeaning)
        # IV2SLS runs both stages internally: 
        # First stage regresses each endog regressors on instruments 
        # Second stage regresses y on the fitted values from the first stage
        
    res = IV2SLS(dependent=y_dm, exog=None, endog=X_dm, instruments=Z_dm) \
            .fit(cov_type="clustered", clusters=clusters)

    # Compact table
    names = [shock, shockL]
    table = pd.DataFrame({
        "coef":       [res.params[n]     for n in names],
        "se (2-way)": [res.std_errors[n] for n in names],
        "t":          [res.tstats[n]     for n in names],
        "p":          [res.pvalues[n]    for n in names],
    }, index=names).round(4)

    print(f"\n=== IV 2SLS (county & year FE absorbed) — {label} ===")
    print(f"Obs (county–years): {int(res.nobs):,}  |  Counties: {ids['fips'].nunique():,}  |  Years: {ids['year'].nunique()}")
    try:
        display(table)
    except NameError:
        print(table.to_string())

    print("Interpretation: β × 0.001 ⇒ $/person per $1,000/person of new 100-mile production.\n")
    return res

# ---- Run both outcomes (current + lag instrumented in both) ----
res_iv_irs_100 = run_distance_iv_strict(
    df_stat2, f"d_d{RADIUS}irswag_cap",
    "IRS wage income per person (Δ), 100-mile IV", RADIUS
)

res_iv_bls_100 = run_distance_iv_strict(
    df_stat2, f"d_d{RADIUS}wages_cap",
    "BLS wage bill per person (Δ), 100-mile IV", RADIUS
)

res_iv_irsother_100 = run_distance_iv_strict(
    df_stat2, f"d_d{RADIUS}irsoth_cap",
    "IRS other income (royalties & nonwage business) per person (Δ), 100-mile IV", RADIUS
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------ FE residualizer ------------------
def twoway_demean_df(df, entity_col="fips", time_col="year", cols=None):
    if cols is None: cols = df.columns # If you didn't pass a list of columns, then operate on all columns
    out = pd.DataFrame(index=df.index) # Prepare an empty output dataframe with the same index as df
    for c in cols: # Loop through each target column; v is the origina series
        v = df[c]
        out[c] = (
            v
            - df.groupby(entity_col)[c].transform("mean")
            - df.groupby(time_col)[c].transform("mean")
            + v.mean()
        )
    return out # Return the residualized dataframe


# ------------------ First-stage scatter (with zoom) ------------------
def first_stage_scatter(
    ax, z, x, annotate=True,
    xlim=None, ylim=None,                     # manual limits
    zoom=None,                                # None | "window" | "percentile"
    z_halfwidth=None, x_halfwidth=None,       # for zoom="window"
    pct=0.98                                   # for zoom="percentile" (0–1)
):
    """
    zoom="window":  set xlim=(-z_halfwidth, z_halfwidth), ylim=(-x_halfwidth, x_halfwidth)
    zoom="percentile": set symmetric limits using percentiles of |Z|, |X| (e.g., pct=0.98)
    Or pass explicit xlim/ylim.
    """
    z = np.asarray(z, float); x = np.asarray(x, float) # Convert inputs to clean float arrays to ensure pure numeric math
    m = np.isfinite(z) & np.isfinite(x); z, x = z[m], x[m] # Keep only pairs where both values are finite (drop NaN/inf consistently)

    # raw points
    ax.scatter(z, x, s=6, alpha=0.10, color="#1f77b4") # Plot the raw/residualized points (tiny dots)

    # Plain bivariate OLS of residualized X on reisdualized Z
    zb, xb = z.mean(), x.mean() # Sample means of z and x
    Szz = np.sum((z - zb)**2) # Total variation in Z
    b1  = np.sum((z - zb)*(x - xb)) / Szz # OLS Slope = Cov(Z,X)/Var(Z)
    a   = xb - b1*zb # Intercept

    # This is the first stage slope of X on Z in a simple regression after residualizing X and Z for county and year FE
    grid = np.linspace(z.min(), z.max(), 200)
    ax.plot(grid, a + b1*grid, color="#1f77b4", linewidth=2)

    # Goodness of fit and strength stats
    xhat = a + b1*z # Fitted values
    ssr = np.sum((xhat - xb)**2); sse = np.sum((x - xhat)**2) # Regression sum of squares and Residual sum of squares
    r2 = ssr / (ssr + sse) if (ssr + sse) > 0 else np.nan # Share of variance in X explained by Z
    se_b1 = np.sqrt((sse / max(1, len(z)-2)) / Szz) 
    t = b1 / se_b1; F = t**2 # T stat for null that b1=0, F stat with one regressor (F=t^2).


    if annotate:
        ax.annotate(
            f"N = {len(z):,}\nslope = {b1:,.3f}\n$R^2$ = {r2:,.3f}\nF = {F:,.1f}",
            xy=(0.02, 0.98), xycoords="axes fraction",
            va="top", ha="left", fontsize=10,
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="0.8")
        ) # Writes the N, R2, and F into box at top left

    # zoom logic
    if xlim is not None: ax.set_xlim(*xlim)
    if ylim is not None: ax.set_ylim(*ylim)

    if zoom == "window":
        if z_halfwidth is not None: ax.set_xlim(-abs(z_halfwidth), abs(z_halfwidth))
        if x_halfwidth is not None: ax.set_ylim(-abs(x_halfwidth), abs(x_halfwidth))
    elif zoom == "percentile":
        zx = np.quantile(np.abs(z), pct)
        xx = np.quantile(np.abs(x), pct)
        ax.set_xlim(-zx, zx); ax.set_ylim(-xx, xx)

    ax.axhline(0, lw=0.8, color="0.35"); ax.axvline(0, lw=0.8, color="0.35")
    ax.grid(True, alpha=0.12)


# ---------- BUILD residualized variables from 100-mile panel ----------
need_100 = [
    "fips", "year",
    "d100newfossilval_cap", "L1_d100newfossilval_cap",            # endogenous regressors (X)
    "d100newfossilval_cap_ins", "L1_d100newfossilval_cap_ins"     # instruments (Z)
]

d100 = df_stat2[need_100].dropna().copy()
for c in need_100:
    if c not in ("fips", "year"):
        d100[c] = pd.to_numeric(d100[c], errors="coerce")
d100 = d100.dropna().copy()

dm100 = twoway_demean_df(
    d100, entity_col="fips", time_col="year",
    cols=[
        "d100newfossilval_cap", "L1_d100newfossilval_cap",
        "d100newfossilval_cap_ins", "L1_d100newfossilval_cap_ins"
    ]
)

Xc100 = dm100["d100newfossilval_cap"].to_numpy()
Zc100 = dm100["d100newfossilval_cap_ins"].to_numpy()
Xl100 = dm100["L1_d100newfossilval_cap"].to_numpy()
Zl100 = dm100["L1_d100newfossilval_cap_ins"].to_numpy()

# ---------- EXAMPLE A: full-range (LAGGED LEFT, CURRENT RIGHT) ----------
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
fig.suptitle("First-stage strength (100-mile panel; county & year FE absorbed)", fontsize=14)

# Lagged on the LEFT
first_stage_scatter(axes[0], Zl100, Xl100)
axes[0].set_title(r"First stage (lagged):  $\tilde{X}_{t-1}$ vs. $\tilde{Z}_{t-1}$")
axes[0].set_xlabel(r"Residualized instrument  $\tilde{Z}_{t-1}$")
axes[0].set_ylabel(r"Residualized production  $\tilde{X}$")

# Current on the RIGHT
first_stage_scatter(axes[1], Zc100, Xc100)
axes[1].set_title(r"First stage (current):  $\tilde{X}_t$ vs. $\tilde{Z}_t$")
axes[1].set_xlabel(r"Residualized instrument  $\tilde{Z}_t$")

plt.tight_layout(); plt.show()

# ---------- EXAMPLE B: zoom near zero (fixed window) ----------
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
fig.suptitle("First-stage (100-mile; zoomed around zero)", fontsize=14)

first_stage_scatter(
    axes[0], Zl100, Xl100, zoom="window",
    z_halfwidth=4e-4, x_halfwidth=8e-4   # adjust if the 100-mile scale differs
)
axes[0].set_title(r"Lagged:  zoomed"); axes[0].set_xlabel(r"$\tilde{Z}_{t-1}$"); axes[0].set_ylabel(r"$\tilde{X}$")

first_stage_scatter(
    axes[1], Zc100, Xc100, zoom="window",
    z_halfwidth=4e-4, x_halfwidth=8e-4
)
axes[1].set_title(r"Current:  zoomed"); axes[1].set_xlabel(r"$\tilde{Z}_t$")

plt.tight_layout(); plt.show()

# ---------- EXAMPLE C: zoom by percentile bands ----------
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
fig.suptitle("First-stage (100-mile; central 98% band)", fontsize=14)

first_stage_scatter(axes[0], Zl100, Xl100, zoom="percentile", pct=0.98)
axes[0].set_title(r"Lagged"); axes[0].set_xlabel(r"$\tilde{Z}_{t-1}$"); axes[0].set_ylabel(r"$\tilde{X}$")

first_stage_scatter(axes[1], Zc100, Xc100, zoom="percentile", pct=0.98)
axes[1].set_title(r"Current"); axes[1].set_xlabel(r"$\tilde{Z}_t$")

plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
from linearmodels.panel import PooledOLS

# ---------------------------------------------------------------------
# Two-way demeaning (absorbs county & year FE). If already defined, skip.
# ---------------------------------------------------------------------
def twoway_demean_df(df, entity_col="fips", time_col="year", cols=None):
    if cols is None: cols = df.columns
    out = pd.DataFrame(index=df.index)
    for c in cols:
        v = df[c]
        out[c] = (
            v
            - df.groupby(entity_col)[c].transform("mean")
            - df.groupby(time_col)[c].transform("mean")
            + v.mean()
        )
    return out

# ---------------------------------------------------------------------
# First-stage tables for 100-mile IV specification (df_stat2)
#   (1)  X_t     ~ Z_t + Z_{t-1}
#   (2)  X_{t-1} ~ Z_t + Z_{t-1}
# FE absorbed via two-way demeaning; 2-way clustered SEs (fips, year).
# ---------------------------------------------------------------------
def first_stage_tables_100mile(df_stat2):
    need = [
        "fips","year",
        "d100newfossilval_cap","L1_d100newfossilval_cap",                  # X_t, X_{t-1}
        "d100newfossilval_cap_ins","L1_d100newfossilval_cap_ins"           # Z_t, Z_{t-1}
    ]
    d = df_stat2[need].dropna().copy()

    # Absorb FE on X and Z via demeaning
    dm = twoway_demean_df(
        d, entity_col="fips", time_col="year",
        cols=[
            "d100newfossilval_cap","L1_d100newfossilval_cap",
            "d100newfossilval_cap_ins","L1_d100newfossilval_cap_ins"
        ]
    )

    def run_one(y_name, Z_names, title):
        # Build working frame (MultiIndex for entity/time so clustering flags work)
        w = pd.concat([d[["fips","year"]], dm[[y_name] + Z_names]], axis=1).dropna()
        w = w.sort_values(["fips","year"]).set_index(["fips","year"])

        y = w[y_name]
        X = w[Z_names]

        res = PooledOLS(y, X).fit(
            cov_type="clustered",
            cluster_entity=True,
            cluster_time=True
        )

        table = pd.DataFrame({
            "coef":       res.params,
            "se (2-way)": res.std_errors,
            "t":          res.tstats,
            "p":          res.pvalues
        }).loc[Z_names].round(4)

        r2 = float(getattr(res, "rsquared", np.nan))

        # Joint Wald test that all instrument coefficients are zero
        f_txt = "n/a"
        try:
            R = np.eye(len(Z_names))
            wald = res.wald_test(R)
            f_txt = f"{wald.stat:.1f} (p={wald.pval:.4f})"
        except Exception:
            try:
                beta = res.params.values.reshape(-1,1)
                V = res.cov.values
                q = len(Z_names)
                from numpy.linalg import pinv
                F = float(beta.T @ pinv(V) @ beta) / q
                f_txt = f"{F:.1f}"
            except Exception:
                pass

        print(f"\n=== First stage — {title} (FE absorbed; 2-way clustered SEs) ===")
        try:
            display(table)
        except NameError:
            print(table.to_string())
        print(f"R^2: {r2:.3f}   |   Joint F (all Z): {f_txt}")

    # (1) Current shock X_t on (Z_t, Z_{t-1})
    run_one(
        y_name="d100newfossilval_cap",
        Z_names=["d100newfossilval_cap_ins","L1_d100newfossilval_cap_ins"],
        title="current shock  X_t  ~  Z_t, Z_{t-1}"
    )

    # (2) Lagged shock X_{t-1} on (Z_t, Z_{t-1})
    run_one(
        y_name="L1_d100newfossilval_cap",
        Z_names=["d100newfossilval_cap_ins","L1_d100newfossilval_cap_ins"],
        title="lagged shock  X_{t-1}  ~  Z_t, Z_{t-1}"
    )

# Run it
first_stage_tables_100mile(df_stat2)

# -- Meaning --
# Coefficients: change in residualized production X per unit change in residualized Z (millions 2014 USD/person).
# SE (2-way): robust to clustering by county (fips) and by year.
# R^2: within share of variance in X explained by Z_t and Z_{t-1}.
# Joint F: joint strength of the instrument set (use this as your weak-IV check).
```

```python
# ============================================================
# Table 2 — County + 100 miles (IV) by industry (lag fix)
# Units: $ per person per $1,000,000/person of new production
# ============================================================

import os, re
import numpy as np
import pandas as pd

# ---------- CONFIG ----------
USER_PATH = "/Users/mickeynguyen/Documents/School /Fracking Paper/Data and Code/Parquet of Data/BLS_Distance_working.parquet"
ALT_PATH  = "/mnt/data/BLS_Distance_working.parquet"

YEARS     = list(range(2005, 2013))  # 2005–2012 inclusive
ID_COL    = "fips"
TIME_COL  = "year"
IND_COL   = "id_industry"

# 100-mile spec variables (from your column list)
Y100   = "d_d100wages_cap"              # Δ wages/salaries per capita (100 miles)
X0_100 = "d100newfossilval_cap"         # current endogenous shock
Z0_100 = "d100newfossilval_cap_ins"     # current instrument

# Report order / labels (paper style)
PAPER_ORDER = [
    (1, "Total"),
    (2, "Mining"),
    (3, "Trans."),
    (4, "Const."),
    (5, "Manuf."),
    (6, "Ed./Health"),
    (7, "Govt."),
    (8, "Oth. serv.")
]

# ---------- HELPERS ----------
def locate_dataset():
    if os.path.exists(USER_PATH): return USER_PATH
    if os.path.exists(ALT_PATH):  return ALT_PATH
    raise FileNotFoundError("Dataset not found. Upload to /mnt/data or adjust USER_PATH/ALT_PATH.")

def balanced_panel(df, group=ID_COL, t=TIME_COL, years=YEARS):
    """Keep counties that have all YEARS (balanced at county–year level)."""
    g = df.groupby(group)[t].agg(["min","max","nunique"])
    ok = g.index[(g["min"] <= years[0]) & (g["max"] >= years[-1]) & (g["nunique"] >= len(years))]
    return df[df[group].isin(ok) & df[t].isin(years)].copy()

def twoway_demean(df, entity=ID_COL, time=TIME_COL, cols=None):
    """Absorb county & year FEs by two-way demeaning."""
    if cols is None: cols = df.columns
    out = pd.DataFrame(index=df.index)
    for c in cols:
        v = pd.to_numeric(df[c], errors="coerce")
        out[c] = v - df.groupby(entity)[c].transform("mean") - df.groupby(time)[c].transform("mean") + v.mean()
    return out

def iv2sls_two_way_cluster(y, X, Z, cluster1, cluster2):
    """
    2SLS with Cameron–Gelbach–Miller two-way clustered SEs.
    y: (n,1), X: (n,k), Z: (n,l); cluster1, cluster2: (n,)
    Returns (beta[k,1], se[k,1]).
    """
    ZTZ = Z.T @ Z
    ZTZ_inv = np.linalg.pinv(ZTZ)
    PzX = Z @ (ZTZ_inv @ (Z.T @ X))
    Pzy = Z @ (ZTZ_inv @ (Z.T @ y))
    XTPZX = X.T @ PzX
    XTPZy = X.T @ Pzy
    XTPZX_inv = np.linalg.pinv(XTPZX)
    beta = XTPZX_inv @ XTPZy
    u = y - X @ beta

    Zu = Z * u
    M0 = (X.T @ Z @ ZTZ_inv) @ (Zu.T @ Zu) @ (ZTZ_inv @ Z.T @ X)

    def meat(keys):
        S = np.zeros((Z.shape[1], Z.shape[1]))
        for key in np.unique(keys):
            idx = (keys == key)
            zsum = (Z[idx, :].T @ u[idx, :]).reshape(-1,1)
            S += zsum @ zsum.T
        return (X.T @ Z @ ZTZ_inv) @ S @ (ZTZ_inv @ Z.T @ X)

    MA = meat(cluster1)
    MB = meat(cluster2)
    V  = XTPZX_inv @ (MA + MB - M0) @ XTPZX_inv
    se = np.sqrt(np.diag(V)).reshape(-1,1)
    return beta, se

# Robust parser for id_industry strings like "1 Total", "Government", etc.
NAME_TO_CODE = {
    "total":1, "mining":2, "transport":3, "trans":3,
    "construct":4, "construction":4,
    "manufact":5, "manufacturing":5, "manuf":5,
    "educ/health":6, "education and health":6, "education health":6, "education":6, "health":6, "ed/health":6,
    "government":7, "govt":7, "gov":7,
    "other services":8, "other service":8, "other serv":8, "oth serv":8, "other":8
}
def parse_id_industry_series(series):
    s = series.astype("string")
    code_from_lead = pd.to_numeric(s.str.extract(r"^\s*(\d+)\s*", expand=False), errors="coerce")
    label_clean = s.str.replace(r"^\s*\d+\s*", "", regex=True).str.strip()
    label_norm = (label_clean.str.lower()
                  .str.replace(r"[^a-z/ ]", "", regex=True)
                  .str.replace(r"\s+", " ", regex=True)
                  .str.strip())
    code_from_name = label_norm.map(NAME_TO_CODE)
    code = code_from_lead.fillna(code_from_name).astype("Int64")
    return code, label_clean

# ---------- LOAD & FILTER ----------
path = locate_dataset()
df = pd.read_parquet(path) if path.endswith(".parquet") else pd.read_feather(path)

# project filters
if "sample" in df.columns:
    df = df[df["sample"] == 1].copy()
df = balanced_panel(df, years=YEARS)

# ---------- LAG FIX: build lags on a county–year table, then merge back ----------
# 1) collapse to one row per (fips, year) for X and Z (they are county–year variables)
base = (df[[ID_COL, TIME_COL, X0_100, Z0_100]]
        .drop_duplicates(subset=[ID_COL, TIME_COL])    # if repeated across industries, keep one
        .sort_values([ID_COL, TIME_COL])
        .copy())

# coerce numeric
for c in [X0_100, Z0_100]:
    base[c] = pd.to_numeric(base[c], errors="coerce")

# 2) compute lags by county over YEAR
base[f"L1_{X0_100}"] = base.groupby(ID_COL)[X0_100].shift(1)
base[f"L1_{Z0_100}"] = base.groupby(ID_COL)[Z0_100].shift(1)

# 3) merge lag columns back to the long industry df
df = df.merge(
    base[[ID_COL, TIME_COL, f"L1_{X0_100}", f"L1_{Z0_100}"]],
    on=[ID_COL, TIME_COL],
    how="left"
)

X1_100 = f"L1_{X0_100}"
Z1_100 = f"L1_{Z0_100}"

# ---------- Parse industries (codes 1..8) ----------
id_code, id_label = parse_id_industry_series(df[IND_COL])
df["id_code"]  = id_code
df["id_label"] = id_label
df = df[df["id_code"].between(1, 8, inclusive="both")].copy()

# ---------- ESTIMATE BY INDUSTRY ----------
results = []
for code, label in PAPER_ORDER:
    w = df[df["id_code"] == code][[ID_COL, TIME_COL, Y100, X0_100, X1_100, Z0_100, Z1_100]].copy()
    # numeric coercion + drop NAs
    for c in [Y100, X0_100, X1_100, Z0_100, Z1_100]:
        w[c] = pd.to_numeric(w[c], errors="coerce")
    w = w.dropna()

    if w.empty:
        results.append({"Industry": label, "coef": np.nan, "se": np.nan, "N": 0})
        continue

    # absorb FE
    dm = twoway_demean(w, entity=ID_COL, time=TIME_COL,
                       cols=[Y100, X0_100, X1_100, Z0_100, Z1_100])
    w_dm = (pd.concat([w[[ID_COL, TIME_COL]], dm], axis=1)
              .dropna()
              .sort_values([ID_COL, TIME_COL]))

    y = w_dm[[Y100]].to_numpy()
    X = w_dm[[X0_100, X1_100]].to_numpy()   # current, lag
    Z = w_dm[[Z0_100, Z1_100]].to_numpy()   # current, lag instruments
    c1 = w_dm[ID_COL].to_numpy()
    c2 = w_dm[TIME_COL].to_numpy()

    beta, se = iv2sls_two_way_cluster(y, X, Z, c1, c2)

    results.append({
        "Industry": label,
        "coef": float(beta[0, 0]),   # coefficient on CURRENT production
        "se":   float(se[0, 0]),     # two-way clustered SE for current-shock coef
        "N":    len(w_dm)
    })

res = pd.DataFrame(results)

# ---------- PRINT LIKE TABLE 2 (one row: +100 miles, IV) ----------
print("\nTABLE 2 — Effects of Fracking on Wages and Salaries by Industry")
print("Specification: County + 100 miles (IV); county & year FE absorbed; two-way clustered SEs\n")

hdr = ["+100 miles (IV)"] + list(res["Industry"].values)
print("{:<18}".format(hdr[0]), end="")
for h in hdr[1:]:
    print("{:>14}".format(h[:12]), end="")
print()

print("{:<18}".format("coef"), end="")
for v in res["coef"].values:
    print("{:>14}".format("" if pd.isna(v) else f"{v:,.0f}"), end="")
print()

print("{:<18}".format("se (cluster)"), end="")
for v in res["se"].values:
    print("{:>14}".format("" if pd.isna(v) else f"{v:,.0f}"), end="")
print("\n")

print("N by industry:")
for _, r in res.iterrows():
    print(f"  {r['Industry']}: {int(r['N']):,}")

print("""
Notes: Lags are computed on a *county–year* table (one row per fips–year) and merged back to the long industry panel.
Dependent variable is Δ wages & salaries per capita within 100 miles (d_d100wages_cap). RHS: current and 1-lag of
d100newfossilval_cap, instrumented by current and 1-lag of d100newfossilval_cap_ins. County & year FEs absorbed
via two-way demeaning; SEs two-way clustered by county and year. Entries are coefficients on CURRENT production
(controlling for the lag), in $/person per $1,000,000/person of new production.
""")
```
